```python
#######################################################################################################
           # RETRIVE DATA FROM JSON FILE AND PRINT IN TABLE
#######################################################################################################

import pandas
with open('data_file.json') as json_file:  
    data = json.load(json_file)
 
text = []
date=[]
like=[]
retw=[]
img=[]
for i in range(0,len(data)):
    key = str(i)
    text.append(data[key]['text'])
    date.append(data[key]['created_at'])
    like.append(data[key]['favorite_count'])
    retw.append(data[key]['retweet_count'])
    count  = 0 
    if 'extended_entities' in data[key].keys():
        #print('yes')
        
        #print(data[key]['extended_entities']['media'])
        for media in data[key]['extended_entities']['media']:
                if media['type'] == 'photo':
                    
                    count += 1
                    #print(count)
    img.append(count)
final_list=[]
final_list.append(text)
final_list.append(date)
final_list.append(like)
final_list.append(retw)
final_list.append(img)

final_list = list(map(list, zip(*final_list)))

headings = ["Text of tweet" , "Date & Time" , "# of likes" , "# of retweets" , "# of images"]

pandas.DataFrame(final_list , columns=headings)   
```
